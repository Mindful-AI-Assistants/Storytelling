```python
## 1- Gr√°fico de PIB per capta - Fundo Preto

# Paleta de cores fortes (voc√™ pode escolher outras combina√ß√µes)
palette = ["#F94144", "#F3722C", "#F8961E", "#90BE6D", "#43AA8B", "#4D5360"]

# Criar a figura e o eixo com fundo preto
plt.figure(figsize=(10, 6))
ax = plt.axes(facecolor='black')  # define o fundo do gr√°fico como preto.

# Criar o gr√°fico de linhas com Seaborn
sns.lineplot(x="year", y="GDP Per Capita (USD)", hue="country", data=countryData,
             palette=palette, linewidth=2.5, ax=ax)  # Aumentar a espessura das linhas

 # Personalizar o gr√°fico
ax.set_title("Evolu√ß√£o do GDP per capita por Pa√≠s", fontsize=20, color='Black')  # T√≠tulo em Prero
ax.set_xlabel("Ano", fontsize=14, color='black')  # R√≥tulos em Preto
ax.set_ylabel("GDP per capita (USD)", fontsize=13, color='Black')
ax.tick_params(axis='x', colors='gray')  # Marcadores do eixo x em cinza
ax.tick_params(axis='y', colors='gray')  # Marcadores do eixo y em gray
ax.grid(True, color='white', alpha=0.4)  # Grade em cinza branco com transparencia

# Ajustar a legenda (opcional)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, facecolor='black',
           edgecolor='None', labelcolor='white')  # Legenda com fundo preto e texto branco

plt.show()


## 2-Evolu√ß√£o temporal do acesso a eletricidade por pa√≠s

# Paleta de cores fortes (voc√™ pode escolher outras combina√ß√µes)
palette = ["#F94144", "#F3722C", "#F8961E", "#90BE6D", "#43AA8B", "#4D5360"]

# Criar a figura e o eixo com fundo preto
plt.figure(figsize=(10, 6))
ax = plt.axes(facecolor='black')  # define o fundo do gr√°fico como preto.

# Criar o gr√°fico de linhas com Seaborn
sns.lineplot(x="year", y="Access to electricity", hue="country", data=countryData,
             palette=palette, linewidth=2.5, ax=ax)

```
